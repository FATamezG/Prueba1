```python
import numpy as np
import pandas as pd
```

```python
datos=pd.read_csv("athlete_events.csv")
datos
```

Ejercicio Básico Punto 1

```python
medallistas=datos[datos["Medal"].notna()]
maximo=medallistas["Age"].max()
medallistas[medallistas["Age"]==maximo]
```

Ejercicio Básico Punto 2

```python
medallas_oro=medallistas[medallistas["Medal"]=="Gold"]
minimo=medallas_oro["Age"].min()
medallas_oro[medallas_oro["Age"]==minimo]
```

Ejercicio Básico Punto 3

```python
maximo_medallista=medallistas["ID"].value_counts().index[0]
salida=datos[datos["ID"]==maximo_medallista]
salida.to_csv('Maximo_Medallista.csv', index=False)
```
