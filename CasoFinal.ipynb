# Alumnos con menor puntuación en Lectura

```python
menor_puntuación=data_students[data_students["reading_score"]==data_students["reading_score"].min()]
scname_gender=menor_puntuación[["school_name","gender"]]
dummies_genero=pd.get_dummies(scname_gender[["gender"]])
plotbygen=pd.concat([scname_gender,dummies_genero],axis=1)
plot_grouped=plotbygen.groupby(["school_name"]).sum()
plot_grouped.plot.bar(figsize=(10,5))
print("Este es un estadístico por escuela y genero para los alumnos con una",
      "\ncalificación en lectura de:"
      ,data_students["reading_score"].min())
```

# Definir una función con dos entradas para generar un estadístico parametrizado por asignatura y calificación

```python
def Func1(asignatura,calificacion):
    filtrado=data_students[data_students[asignatura]==calificacion]
    scname_gender=filtrado[["school_name","gender"]]
    dummies_genero=pd.get_dummies(scname_gender[["gender"]])
    plotbygen=pd.concat([scname_gender,dummies_genero],axis=1)
    plot_grouped=plotbygen.groupby(["school_name"]).sum()
    plot_grouped.plot.bar(figsize=(10,5))
    print("Este es un estadístico por escuela y género para los alumnos con una",
          "\ncalificación en {0} de: {1}".format(asignatura,calificacion))
```

```python
Func1("reading_score",63)
```

```python
Func1("reading_score",75)
```

```python
Func1("math_score",90)
```

# Definir una función que obtenga en base a la función anterior los 4 estadísticos importantes: alúmnos por género con mayor y menor calificación en Reading y en Math

```python
def Func2():
    Func1("reading_score",data_students["reading_score"].max())
    Func1("reading_score",data_students["reading_score"].min())
    Func1("math_score",data_students["math_score"].max())
    Func1("math_score",data_students["math_score"].min())    
```
